## 1. Construcci√≥n de circuitos cu√°nticos

Antes de poder aplicar operaciones a los qubits, necesitamos colocarlos en una estructura que permita manipularlos: el **circuito cu√°ntico**.

Un circuito cu√°ntico es un modelo fundamental de la computaci√≥n cu√°ntica. Al igual que los circuitos cl√°sicos, representa una secuencia de operaciones sobre qubits: compuertas cu√°nticas, mediciones, inicializaciones y m√°s.  


üìå Visualmente, un circuito cu√°ntico se representa as√≠:

- Las l√≠neas **horizontales** son **qubits**.
- Las l√≠neas **dobles** son **bits cl√°sicos**.
- El **tiempo avanza** de izquierda a derecha.
- Las **compuertas** se colocan sobre los qubits a lo largo de la l√≠nea horizontal.

## üîß 2. Visualizaci√≥n del circuito

Veamos c√≥mo incluir en un circuito uno de los qubits que creamos anteriormente. Usamos la clase `QuantumCircuit` de Qiskit para construir el circuito y luego lo visualizamos gr√°ficamente.


```python
from qiskit import QuantumCircuit

circuit = QuantumCircuit(1)  # Crear un circuito cu√°ntico con 1 qubit

circuit.draw('mpl')          # Mostramos la representaci√≥n gr√°fica del circuito
```

## ‚öôÔ∏è Circuito con M√∫ltiples Qubits y Bits Cl√°sicos

En el siguiente ejemplo se define un circuito cu√°ntico con **3 qubits** y **2 bits cl√°sicos**.  
No es necesario crear los qubits usando `QuantumRegister`, basta con indicar el n√∫mero de qubits y bits cl√°sicos al inicializar el circuito.

- El **qubit 0** se mide en el **bit cl√°sico 1**.
- El **qubit 2** se mide en el **bit cl√°sico 0**.

M√°s adelante analizaremos el efecto de estas operaciones.


```python
from qiskit import QuantumCircuit

circ = QuantumCircuit(3, 2)   # Circuito con 3 qubits y 2 bits cl√°sicos

circ.h(0)           # Aplicar compuerta Hadamard al qubit 0
circ.cx(0, 1)       # Aplicar compuerta CNOT entre qubit 0 (control) y 1 (objetivo)
circ.cx(0, 2)       # Aplicar compuerta CNOT entre qubit 0 (control) y 2 (objetivo)

circ.measure(0, 1)  # Medir el qubit 0 y guardar el resultado en el bit cl√°sico 1
circ.measure(2, 0)  # Medir el qubit 2 y guardar el resultado en el bit cl√°sico 0

circ.draw('mpl')    # Mostrar el circuito gr√°ficamente
```

### ‚ú® Compuerta Hadamard ‚Äì Superposici√≥n Cu√°ntica

La compuerta **Hadamard (H)** transforma un qubit desde un estado base (`|0‚ü©` o `|1‚ü©`) a un estado de **superposici√≥n**.

üîπ Esto significa que el qubit **tendr√° amplitudes tanto para `|0‚ü©` como para `|1‚ü©` al mismo tiempo**:

- `H |0‚ü© = (|0‚ü© + |1‚ü©) / ‚àö2`
- `H |1‚ü© = (|0‚ü© - |1‚ü©) / ‚àö2`

üß† Es una de las compuertas m√°s importantes en computaci√≥n cu√°ntica, ya que permite **explorar m√∫ltiples caminos al mismo tiempo** en un algoritmo.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_bloch_multivector
from IPython.display import display

# Crear circuito con qubit en |0‚ü©
circ_h = QuantumCircuit(1)
psi_0 = Statevector(circ_h)  # Estado antes de aplicar H

# Aplicar compuerta Hadamard
circ_h.h(0)
psi_1 = Statevector(circ_h)  # Estado despu√©s de aplicar H

# Mostrar circuito
display(circ_h.draw('mpl'))

# Mostrar estado inicial
print("üëâ Estado inicial:")
print("‚û§ Vector:")
display(array_to_latex(psi_0.data))
print("‚û§ Ket:")
display(psi_0.draw('latex'))

# Mostrar estado despu√©s de Hadamard
print("\nüëâ Estado despu√©s de aplicar Hadamard:")
print("‚û§ Vector:")
display(array_to_latex(psi_1.data))
print("‚û§ Ket:")
display(psi_1.draw('latex'))
```

```python
# Esfera de Bloch antes y despu√©s
print("üîµ Estado inicial:")
display(plot_bloch_multivector(psi_0))

print("üîµ Estado despu√©s de aplicar Hadamard:")
display(plot_bloch_multivector(psi_1))
```

üìå **Nota**: La compuerta Hadamard es fundamental porque **saca al qubit de la base cl√°sica** (`|0‚ü©` o `|1‚ü©`) y lo lleva a una **superposici√≥n**, permitiendo que el sistema cu√°ntico explore m√∫ltiples caminos al mismo tiempo. 

Es el punto de partida en la mayor√≠a de los algoritmos cu√°nticos.


### üîÅ Compuerta X ‚Äì Inversi√≥n del Qubit

Aplicamos la compuerta **X** (tambi√©n conocida como **NOT cu√°ntica**) al √∫nico qubit del circuito.

üîπ Esta operaci√≥n invierte el estado del qubit, es decir:
- Transforma el estado `|0‚ü©` en `|1‚ü©`
- Y `|1‚ü©` en `|0‚ü©`

üß† Es an√°loga a una compuerta **NOT cl√°sica**, pero aplicada en el contexto de la computaci√≥n cu√°ntica.


```python
from qiskit import QuantumCircuit

circ1 = QuantumCircuit(1)  # Circuito con 1 qubit

circ1.x(0)                 # Aplicar compuerta X al qubit 0

circ1.draw('mpl')          # Mostrar el circuito
```

### Obtener el vector de estado con Qiskit

Con la clase `Statevector` de Qiskit, se puede obtener el estado de un circuito en el punto de se desee:

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex


circ1 = QuantumCircuit(1)    # circuito con 1 qubit

psi_0 = Statevector(circ1)   # estado justo despu√©s de crear el circuito

circ1.x(0)                   # aplicar compuerta X al (√∫nico) qubit 0

psi_1 = Statevector(circ1)   # estado despu√©s aplicar la compuerta X

display(circ1.draw('mpl'))   # mostrar el circuito

print("üëâ Estado inicial:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_0.data))
print("‚û§ en forma de ket:")
display(psi_0.draw('latex'))
print()
print("üëâ Estado despu√©s de X:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_1.data))
print("‚û§ en forma de ket:")
display(psi_1.draw('latex'))
```

### üîç Visualizaci√≥n en la Esfera de Bloch

La siguiente visualizaci√≥n muestra el estado del qubit **antes** y **despu√©s** de aplicar la compuerta **X** en la esfera de Bloch.

Recordemos que:

- El estado inicial `|0‚ü©` se representa en el polo norte de la esfera.
- La compuerta **X** transforma `|0‚ü©` en `|1‚ü©`, que se ubica en el polo sur de la esfera.

Con la funci√≥n `plot_bloch_multivector()` podemos visualizar un qubit, indicando el vector de estado.


```python

# üîµ Estado inicial
psi_0 = Statevector.from_label('0')
print("üîµ Estado inicial:")
display(plot_bloch_multivector(psi_0))

# Aplicar compuerta X
qc = QuantumCircuit(1)
qc.x(0)
psi_1 = Statevector.from_instruction(qc)

# üîµ Estado despu√©s de X
print("üîµ Estado despu√©s de aplicar X:")
display(plot_bloch_multivector(psi_1))

```

### üîÅ Compuerta Y ‚Äì Rotaci√≥n en el Eje Y

Aplicamos la compuerta **Y** al √∫nico qubit del circuito.

üîπ Esta operaci√≥n realiza una rotaci√≥n de 180¬∞ alrededor del eje **Y** de la esfera de Bloch.

- Transforma el estado `|0‚ü©` en `i|1‚ü©`
- Y `|1‚ü©` en `-i|0‚ü©`

üß† A diferencia de la compuerta X (que solo invierte), la compuerta **Y** tambi√©n introduce una **fase imaginaria**, por lo tanto tiene un efecto m√°s complejo en el estado cu√°ntico.


```python
from qiskit import QuantumCircuit

circ_y = QuantumCircuit(1)  # Circuito con 1 qubit

circ_y.y(0)                 # Aplicar compuerta Y al qubit 0

circ_y.draw('mpl')          # Mostrar el circuito
```

### Obtener el vector de estado con Qiskit

Con la clase `Statevector` de Qiskit, se puede obtener el estado de un circuito en el punto que se desee:


```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex

circ_y = QuantumCircuit(1)

psi_y_0 = Statevector(circ_y)  # estado inicial antes de aplicar Y

circ_y.y(0)                    # aplicar compuerta Y

psi_y_1 = Statevector(circ_y)  # estado despu√©s de aplicar Y

display(circ_y.draw('mpl'))

print("üëâ Estado inicial:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_y_0.data))
print("‚û§ en forma de ket:")
display(psi_y_0.draw('latex'))
print()
print("üëâ Estado despu√©s de Y:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_y_1.data))
print("‚û§ en forma de ket:")
display(psi_y_1.draw('latex'))
```

### üîç Visualizaci√≥n en la Esfera de Bloch

La siguiente visualizaci√≥n muestra el estado del qubit **antes** y **despu√©s** de aplicar la compuerta **Y** en la esfera de Bloch.

Con la funci√≥n `plot_bloch_multivector()` podemos visualizar el estado cu√°ntico de forma gr√°fica:


```python
from qiskit.visualization import plot_bloch_multivector
from IPython.display import display

# üîµ Estado inicial
psi_y_0 = Statevector.from_label('0')
print("üîµ Estado inicial:")
display(plot_bloch_multivector(psi_y_0))

# Aplicar compuerta Y
qc_y = QuantumCircuit(1)
qc_y.y(0)
psi_y_1 = Statevector.from_instruction(qc_y)

# üîµ Estado despu√©s de Y
print("üîµ Estado despu√©s de aplicar Y:")
display(plot_bloch_multivector(psi_y_1))
```

üìå **Nota**: Aunque la compuerta **Y** y la compuerta **X** transforman el estado `|0‚ü©` en `|1‚ü©`, la **compuerta Y introduce una fase imaginaria** (`i|1‚ü©`), lo cual **no es visible en la esfera de Bloch**, ya que esta representaci√≥n solo muestra componentes reales del vector de estado. Por eso, **la esfera de Bloch se ver√° igual** que con la compuerta **X**, aunque el estado cu√°ntico completo sea distinto.


### üîÅ Compuerta Z ‚Äì Cambio de Fase

La compuerta **Z** realiza una rotaci√≥n de 180¬∞ alrededor del eje **Z** de la esfera de Bloch. 

üîπ Esta operaci√≥n **no altera el estado `|0‚ü©`**, pero s√≠ cambia el signo de `|1‚ü©`:

- `|0‚ü©` ‚Üí `|0‚ü©`
- `|1‚ü©` ‚Üí `‚àí|1‚ü©`

üß† A nivel de probabilidad de medici√≥n, el estado sigue pareciendo el mismo, pero **la fase relativa cambia**, lo cual es **crucial en algoritmos cu√°nticos que dependen de interferencia**.

---

Para evidenciar este cambio, primero **iniciamos el qubit en `|1‚ü©`** aplicando una compuerta **X**, y **luego aplicamos la compuerta Z**.



```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_bloch_multivector
from IPython.display import display

# Crear circuito y preparar el qubit en |1‚ü©
circ3 = QuantumCircuit(1)
circ3.x(0)                   # estado |1‚ü©

psi_0 = Statevector(circ3)   # estado antes de aplicar Z

circ3.z(0)                   # aplicar compuerta Z

psi_1 = Statevector(circ3)   # estado despu√©s de Z

# Dibujar circuito
display(circ3.draw('mpl'))

# Mostrar estados
print("üëâ Estado inicial:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_0.data))
print("‚û§ en forma de ket:")
display(psi_0.draw('latex'))
print()
print("üëâ Estado despu√©s de Z:")
print("‚û§ en forma de vector:")
display(array_to_latex(psi_1.data))
print("‚û§ en forma de ket:")
display(psi_1.draw('latex'))

# Visualizaci√≥n en Bloch
print()
print("üîµ Estado inicial:")
display(plot_bloch_multivector(psi_0))

print("üîµ Estado despu√©s de Z:")
display(plot_bloch_multivector(psi_1))
```

üìå **Nota**: Cuando el qubit est√° en el estado `|0‚ü©`, la compuerta **Z** no produce ning√∫n cambio visible.

Sin embargo, cuando est√° en `|1‚ü©`, el efecto de la compuerta **Z** es **invertir su fase** a `‚àí|1‚ü©`, lo que se refleja como un giro en la esfera de Bloch desde el polo sur hacia el otro lado del eje Z.

Esto **demuestra visualmente** que la compuerta Z s√≠ tiene un efecto real, aunque no cambie probabilidades de medici√≥n.


‚ùì **¬øPor qu√© las dos esferas se ven iguales si el estado cambi√≥ de `|1‚ü©` a `‚àí|1‚ü©`?**

Aunque `|1‚ü©` y `‚àí|1‚ü©` **son estados diferentes matem√°ticamente**, en la esfera de Bloch **representan el mismo punto**.

üìå Esto se debe a que la esfera de Bloch **no puede representar fases globales** (como un signo negativo o un factor imaginario com√∫n en todo el estado). La visualizaci√≥n solo muestra diferencias **relativas** entre los componentes del qubit, no absolutas.

üß† En otras palabras:  
- `|1‚ü©` y `‚àí|1‚ü©` tienen la **misma probabilidad de medici√≥n**
- Pero pueden producir **interferencias diferentes** si se combinan con otros qubits

Por eso, aunque `Z` s√≠ modifica el estado, la esfera de Bloch **parece no cambiar**.
