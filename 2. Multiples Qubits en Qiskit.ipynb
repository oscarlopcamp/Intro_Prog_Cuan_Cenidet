## üß© Secci√≥n 1 ‚Äì Creaci√≥n de sistemas con m√∫ltiples qubits

Hasta ahora hemos trabajado con circuitos de **un solo qubit**, donde el estado era una combinaci√≥n de `|0‚ü©` y `|1‚ü©`.

Ahora construiremos circuitos con **dos qubits**, lo cual da lugar a **cuatro posibles estados base**:

- `|00‚ü©`
- `|01‚ü©`
- `|10‚ü©`
- `|11‚ü©`

Estos forman la **base computacional** para sistemas de 2 qubits.  
Cualquier combinaci√≥n lineal de estos estados es v√°lida para describir un sistema cu√°ntico de dos qubits.

---

En Qiskit, se crea un sistema de 2 qubits as√≠:



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from IPython.display import display

# Crear circuito con 2 qubits
qc = QuantumCircuit(2)
psi = Statevector(qc)

# Mostrar circuito vac√≠o
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema |00‚ü©:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))
````

---

### ‚úÖ Explicaci√≥n del resultado

- El sistema comienza en el estado `|00‚ü©`, lo cual significa que **ambos qubits est√°n en `|0‚ü©`**.
- El vector de estado tiene **cuatro componentes**, y solo la correspondiente a `|00‚ü©` tiene valor `1`; las dem√°s son `0`.
- El orden en Qiskit sigue la convenci√≥n **little-endian**, lo que significa que el qubit **de mayor √≠ndice (qubit 1)** es el m√°s significativo.

Esto es importante para interpretar correctamente las posiciones en el vector de estado:

$$
[|00‚ü©,\ |01‚ü©,\ |10‚ü©,\ |11‚ü©]
$$

---

üìå En la siguiente secci√≥n aplicaremos compuertas a uno de los qubits para cambiar el estado, y observaremos c√≥mo se modifica el sistema completo.

# ‚öôÔ∏è Secci√≥n 2 ‚Äì Aplicaci√≥n de compuertas individuales en sistemas de 2 qubits

Ahora que hemos creado un sistema con dos qubits en estado `|00‚ü©`, aplicaremos una **compuerta X al qubit 1**, para transformar el estado a `|01‚ü©`.

Recordemos que:
- `qubit 0` es el de **abajo** (menos significativo)
- `qubit 1` es el de **arriba** (m√°s significativo)

Por tanto, aplicar `X(1)` transforma:

$$
|00‚ü© \quad \longrightarrow \quad |01‚ü©
$$

Vamos a observar c√≥mo se modifica el circuito, el vector de estado y su notaci√≥n ket.

Tambi√©n utilizaremos la funci√≥n `plot_state_qsphere()` para visualizar el estado resultante en una **QSphere**, una herramienta gr√°fica que representa el **estado completo del sistema cu√°ntico**, incluyendo las **probabilidades y fases cu√°nticas** de cada uno de los estados base.



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar compuerta X al qubit 1
qc = QuantumCircuit(2)
qc.x(1)
psi = Statevector(qc)

# Mostrar circuito con compuerta X
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)

````

### üßø ¬øQu√© estamos viendo en la QSphere?

La **QSphere** es una herramienta gr√°fica que nos permite observar el estado cu√°ntico completo del sistema, mostrando:

- Los **estados base** posibles (como `|00‚ü©`, `|01‚ü©`, `|10‚ü©`, `|11‚ü©`)
- Las **probabilidades** de cada uno (tama√±o del punto)
- Las **fases cu√°nticas** relativas (color del punto)

En esta visualizaci√≥n:

- El punto correspondiente a `|11‚ü©` es el √∫nico que aparece grande, lo que indica que el sistema est√° **completamente en ese estado**.
- Los dem√°s puntos (`|00‚ü©`, `|01‚ü©`, `|10‚ü©`) tienen **amplitud cero**, por lo que no aparecen visibles.
- El **color del punto** indica la **fase cu√°ntica** del estado. En este caso, al no haber superposici√≥n, no hay interferencia ni fases relativas importantes.

üìå La QSphere es especialmente √∫til para representar **estados con m√∫ltiples componentes y fases complejas**, como cuando hay **superposici√≥n o entrelazamiento**. Aqu√≠, como solo hay un estado (`|11‚ü©`), se ve un solo punto grande y bien definido.



---

### ‚úÖ Explicaci√≥n del resultado

Al aplicar la compuerta **X al qubit 1**, el estado del sistema cambia de:

$$
|00‚ü© \quad \longrightarrow \quad |01‚ü©
$$

Esto significa que ahora:
- El **qubit 1 (m√°s significativo)** est√° en `|1‚ü©`
- El **qubit 0 (menos significativo)** permanece en `|0‚ü©`

Por tanto, el sistema est√° en el estado `|01‚ü©`.

üìå En el vector de estado, la √∫nica amplitud distinta de cero est√° en la **segunda posici√≥n**, que representa `|01‚ü©` seg√∫n el orden de Qiskit:

$$
[|00‚ü©,\ |01‚ü©,\ |10‚ü©,\ |11‚ü©]
$$


### ‚úÖ Explicaci√≥n del resultado

## ‚öôÔ∏è Secci√≥n 3 ‚Äì Aplicaci√≥n de compuertas individuales en ambos qubits

Hasta ahora hemos aplicado compuertas individuales a **un solo qubit a la vez**.  
En esta secci√≥n aplicaremos **una compuerta a cada qubit** dentro del mismo circuito.

Como ejemplo, aplicaremos la compuerta **X al qubit 0** y la compuerta **X al qubit 1**, lo cual transformar√° el estado inicial `|00‚ü©` en `|11‚ü©`.

Esto nos permitir√° observar c√≥mo se comporta el sistema cuando **ambos qubits cambian simult√°neamente**.

---




````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar compuertas X a ambos qubits
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)

````

### ‚úÖ Explicaci√≥n del resultado

$$
|00‚ü© \quad \longrightarrow \quad |11‚ü©
$$

Esto significa que ahora:
- El **qubit 1 (m√°s significativo)** est√° en `|1‚ü©`
- El **qubit 0 (menos significativo)** tambi√©n est√° en `|1‚ü©`

Por tanto, el sistema est√° en el estado `|11‚ü©`.

üìå En el vector de estado, la √∫nica amplitud distinta de cero est√° en la **cuarta posici√≥n**, que representa `|11‚ü©` seg√∫n el orden de Qiskit:

$$
[|00‚ü©,\ |01‚ü©,\ |10‚ü©,\ |11‚ü©]
$$

## ‚öôÔ∏è Secci√≥n 4 ‚Äì Superposici√≥n en sistemas de 2 qubits

Hasta ahora hemos trabajado con qubits en estados cl√°sicos (`|0‚ü©`, `|1‚ü©`).  
En esta secci√≥n vamos a introducir el concepto de **superposici√≥n**, aplicando una **compuerta Hadamard** al **qubit 0** mientras que el qubit 1 permanece en `|0‚ü©`.

Esto genera un estado donde el sistema cu√°ntico est√° simult√°neamente en `|00‚ü©` y `|10‚ü©`.

Utilizaremos la **QSphere** para visualizar c√≥mo se distribuyen las amplitudes y las fases cu√°nticas en un sistema de 2 qubits en superposici√≥n.

---


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar Hadamard al qubit 0
qc = QuantumCircuit(2)
qc.h(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)
````

### üß† Interpretaci√≥n de la QSphere

En esta visualizaci√≥n podemos ver:

- Dos puntos grandes: `|00‚ü©` y `|10‚ü©`, lo que indica que el sistema est√° en superposici√≥n entre esos dos estados
- Ambos tienen **amplitud 1/‚àö2** y la **misma fase** (por eso el mismo color)
- Los estados `|01‚ü©` y `|11‚ü©` no aparecen (amplitud 0)

Esto representa el estado:

$$
|\psi‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |10‚ü©)
$$

üìå Es decir: el **qubit 0** est√° en superposici√≥n, y el **qubit 1** sigue en `|0‚ü©`.

## ‚öôÔ∏è Secci√≥n 5 ‚Äì Entrelazamiento: creaci√≥n del estado de Bell

Ahora aplicaremos una compuerta **CNOT** al sistema cu√°ntico que generamos previamente.  
Este sistema estaba en superposici√≥n: `(|00‚ü© + |10‚ü©)/‚àö2`, es decir, el **qubit 0** en superposici√≥n y el **qubit 1** en `|0‚ü©`.

La **CNOT** aplicar√° una compuerta X al qubit 1 **solo si el qubit 0 est√° en `|1‚ü©`**.

---

### üîó ¬øQu√© significa que est√©n entrelazados?

Este estado es **entrelazado**, lo que significa que los qubits est√°n conectados:

- Si mides el qubit 0 y da `0`, el qubit 1 tambi√©n ser√° `0`
- Si mides el qubit 0 y da `1`, el qubit 1 tambi√©n ser√° `1`

Aunque no los mires al mismo tiempo, **los resultados siempre estar√°n correlacionados**.

Este tipo de conexi√≥n cu√°ntica es lo que hace al entrelazamiento tan especial y √∫til en la computaci√≥n cu√°ntica.

---

En Qiskit, esta compuerta se codifica como:

```python
qc.cx(control, target)



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito: Hadamard en qubit 0, luego CNOT (0 controla a 1)
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)
````

### üß† ¬øQu√© est√° pasando en este circuito?

Vamos a explicar paso a paso, en palabras simples, qu√© ocurre en este circuito con una compuerta Hadamard y una CNOT:

---

#### 1. Estado inicial

Ambos qubits empiezan en `|0‚ü©`, por lo tanto el sistema est√° en:
$$
|00‚ü©
$$

---

#### 2. Aplicamos Hadamard al qubit 0

Esto pone al **qubit 0** en **superposici√≥n**: est√° a la vez en `|0‚ü©` y `|1‚ü©`.  
El qubit 1 sigue en `|0‚ü©`.

El nuevo estado del sistema es:

$$
|\psi‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |10‚ü©)
$$

Esto significa que hay:
- 50% de probabilidad de que est√©n en `|00‚ü©`
- 50% de probabilidad de que est√©n en `|10‚ü©`

---

#### 3. Aplicamos la compuerta CNOT

La **CNOT** funciona as√≠:
- Si el **qubit 0 (control)** est√° en `0`, no hace nada
- Si est√° en `1`, aplica una compuerta **X** al **qubit 1 (objetivo)**

Entonces:
- `|00‚ü©` se queda igual
- `|10‚ü©` se transforma en `|11‚ü©`

El nuevo estado es:

$$
|\Phi^+‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |11‚ü©)
$$


---



### üîó Interpretaci√≥n del estado de Bell

El resultado de este circuito es un **estado entrelazado**:

$$
|\Phi^+‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |11‚ü©)
$$

Esto significa que:

- Si medimos el sistema y el **qubit 0 da `0`**, el **qubit 1 tambi√©n ser√° `0`**
- Si el **qubit 0 da `1`**, el **qubit 1 tambi√©n ser√° `1`**
- ¬°Est√°n perfectamente correlacionados! üîó

En la **QSphere** se observa:

- Dos puntos grandes: `|00‚ü©` y `|11‚ü©`
- Ambos tienen amplitud 1/‚àö2
- El **color diferente** indica que tienen una **fase relativa** entre s√≠ (aunque en este caso puede ser la misma)

üìå Este tipo de estados son la base para muchas aplicaciones de la computaci√≥n y criptograf√≠a cu√°ntica.

## ‚öôÔ∏è Secci√≥n 6 ‚Äì Aplicar m√∫ltiples compuertas en un mismo qubit

Hasta ahora solo hemos aplicado **una compuerta por qubit**.  
En esta secci√≥n aprenderemos que es posible aplicar **varias compuertas en el mismo qubit, una tras otra**.

Esto es muy com√∫n en circuitos cu√°nticos, y permite **modificar gradualmente el estado de un qubit**.

Veremos dos casos:

- Aplicar `H` seguido de `X` al qubit 0
- Aplicar `X` seguido de `H` al mismo qubit

Veremos que **el orden importa** y el estado final cambia dependiendo del orden.

---


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Caso 1: aplicar Hadamard luego X al qubit 0
qc = QuantumCircuit(1)
qc.h(0)
qc.x(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø QSphere del estado:")
plot_state_qsphere(psi)
````

### ‚ú® Caso 1 ‚Äì Aplicar `H` seguido de `X` al qubit 0

Primero aplicamos **Hadamard**, que pone al qubit 0 en superposici√≥n:

$$
|0‚ü© \longrightarrow \frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©)
$$

Despu√©s aplicamos la compuerta **X**, que **intercambia `|0‚ü©` por `|1‚ü©` y viceversa**.

El efecto total es:

$$
\frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©) \longrightarrow \frac{1}{\sqrt{2}}(|1‚ü© + |0‚ü©) = \frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©)
$$

En este caso, **el estado final es el mismo**, pero eso **no siempre ocurre**.  
Veamos qu√© pasa si invertimos el orden.

---


````python
# Caso 2: aplicar X luego Hadamard al qubit 0
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø QSphere del estado:")
plot_state_qsphere(psi)
````

### üîÅ Caso 2 ‚Äì Aplicar `X` seguido de `H` al qubit 0

Ahora primero aplicamos la compuerta **X**, que cambia el estado de `|0‚ü©` a `|1‚ü©`:

$$
|0‚ü© \longrightarrow |1‚ü©
$$

Luego aplicamos la compuerta **Hadamard** al qubit 0, y transforma:

$$
|1‚ü© \longrightarrow \frac{1}{\sqrt{2}}(|0‚ü© - |1‚ü©)
$$

Este estado final **no es el mismo que en el caso anterior**, ya que ahora hay una **fase negativa**:

$$
|\psi‚ü© = \frac{1}{\sqrt{2}}(|0‚ü© - |1‚ü©)
$$

üìå Esto demuestra que **el orden de las compuertas importa** y afecta al resultado final.

## ‚öôÔ∏è Secci√≥n 7 ‚Äì Generaci√≥n de los cuatro estados de Bell

Los **estados de Bell** son ejemplos fundamentales de **entrelazamiento cu√°ntico**.  
Cada uno representa un sistema de dos qubits **perfectamente correlacionados**, con distintas combinaciones de fases y amplitudes.

Los 4 estados de Bell son:

- $|\Phi^+‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |11‚ü©)$  
- $|\Phi^-‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© - |11‚ü©)$  
- $|\Psi^+‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© + |10‚ü©)$  
- $|\Psi^-‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© - |10‚ü©)$

En esta secci√≥n construiremos **cada uno paso a paso**, usando compuertas simples y visualizando su resultado.

---



---

### üî∑ Estado de Bell 1: $|\Phi^+‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© + |11‚ü©)$

Este es el estado de Bell m√°s com√∫n.  
Se genera aplicando una **Hadamard** al qubit 0 para crear superposici√≥n, y luego una compuerta **CNOT** para copiar esa superposici√≥n al qubit 1 y as√≠ entrelazarlos.

Compuertas usadas:
- `H(0)`: pone el qubit 0 en superposici√≥n
- `CNOT(0, 1)`: entrelaza los qubits

---


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Circuito para el estado |Œ¶‚Å∫‚ü©
qc = QuantumCircuit(2)
qc.h(0)       # Hadamard en qubit 0
qc.cx(0, 1)   # CNOT: control=0, target=1

# Obtener el estado
psi = Statevector(qc)

# Visualizaci√≥n
display(qc.draw('mpl'))
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))
print("üß™ Notaci√≥n ket:")
display(psi.draw('latex'))
print("üßø QSphere:")
plot_state_qsphere(psi)
````

---

### üî∑ Estado de Bell 2: $|\Phi^-‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© - |11‚ü©)$

Este estado es similar a $|\Phi^+‚ü©$, pero incluye una **fase negativa** en el t√©rmino $|11‚ü©$.

Para generarlo, usamos:

- `H(0)`: crea la superposici√≥n
- `Z(0)`: agrega una fase negativa al estado $|1‚ü©$
- `CNOT(0, 1)`: entrelaza los qubits

üìå **Compuerta nueva: `Z`**  
La compuerta Z **no cambia la probabilidad**, pero **agrega una fase negativa a $|1‚ü©$**.  
Esto no se nota en mediciones directas, pero **s√≠ cambia el estado cu√°ntico**.

---


````python
# Circuito para el estado |Œ¶‚Åª‚ü©
qc = QuantumCircuit(2)
qc.h(0)       # Hadamard en qubit 0
qc.z(0)       # Fase negativa a |1‚ü©
qc.cx(0, 1)   # Entrelazamiento con CNOT

# Obtener el estado
psi = Statevector(qc)

# Visualizaci√≥n
display(qc.draw('mpl'))
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))
print("üß™ Notaci√≥n ket:")
display(psi.draw('latex'))
print("üßø QSphere:")
plot_state_qsphere(psi)
````

### üß† Paso a paso ‚Äì C√≥mo se forma $|\Phi^-‚ü©$

1. **Estado inicial:**
   $$|00‚ü©$$

2. **Hadamard en qubit 0:**
   El qubit 0 entra en superposici√≥n:
   $$
   |0‚ü© \longrightarrow \frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©)
   $$

   Como el qubit 1 sigue en $|0‚ü©$, el sistema ahora est√° en:
   $$
   \frac{1}{\sqrt{2}}(|00‚ü© + |10‚ü©)
   $$

3. **Z en qubit 0:**
   Agrega una **fase negativa** a la parte $|1‚ü©$ del qubit 0:
   $$
   \frac{1}{\sqrt{2}}(|00‚ü© + |10‚ü©) \longrightarrow \frac{1}{\sqrt{2}}(|00‚ü© - |10‚ü©)
   $$

4. **CNOT:**
   Si qubit 0 = `1`, cambia qubit 1. Entonces:
   - $|00‚ü©$ se queda igual
   - $|10‚ü©$ se convierte en $|11‚ü©$

   Resultado:
   $$
   |\Phi^-‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© - |11‚ü©)
   $$

---

üìå En la **QSphere**, se observan dos puntos: $|00‚ü©$ y $|11‚ü©$ con **colores opuestos**, indicando la diferencia de fase.

---



---

### üî∑ Estado de Bell 3: $|\Psi^+‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© + |10‚ü©)$

Este estado representa una **superposici√≥n entre dos configuraciones opuestas**:
- qubit 0 = `0` y qubit 1 = `1`
- qubit 0 = `1` y qubit 1 = `0`

Es un estado **entrelazado**, donde si medimos uno de los qubits, el otro estar√° siempre en el valor contrario.

Para generarlo, usamos:

- `X(1)`: cambia el qubit 1 de `|0‚ü©` a `|1‚ü©`
- `H(0)`: pone el qubit 0 en superposici√≥n
- `CNOT(0, 1)`: entrelaza los qubits

üìå **Compuerta nueva: `X(1)`**  
La compuerta X invierte el estado del qubit:
- `|0‚ü©` ‚Üí `|1‚ü©`
- `|1‚ü©` ‚Üí `|0‚ü©`

---


````python
# Circuito para el estado |Œ®‚Å∫‚ü©
qc = QuantumCircuit(2)
qc.x(1)       # 1. Invertir qubit 1: |00‚ü© ‚Üí |01‚ü©
qc.h(0)       # 2. Superposici√≥n en qubit 0
qc.cx(0, 1)   # 3. Entrelazamiento con CNOT

# Obtener el estado
psi = Statevector(qc)

# Visualizaci√≥n
display(qc.draw('mpl'))
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))
print("üß™ Notaci√≥n ket:")
display(psi.draw('latex'))
print("üßø QSphere:")
plot_state_qsphere(psi)
````

### üß† Paso a paso ‚Äì C√≥mo se forma $|\Psi^+‚ü©$

1. **Estado inicial:**
   $$
   |00‚ü©
   $$

2. **Aplicamos `X(1)`**:
   Esto cambia el qubit 1 de $|0‚ü©$ a $|1‚ü©$:
   $$
   |00‚ü© \longrightarrow |01‚ü©
   $$

3. **Aplicamos `H(0)`**:
   Esto pone el qubit 0 en superposici√≥n:
   $$
   |0‚ü© \longrightarrow \frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©)
   $$

   Entonces:
   $$
   \frac{1}{\sqrt{2}}(|0‚ü© + |1‚ü©) \otimes |1‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© + |11‚ü©)
   $$

4. **Aplicamos `CNOT(0, 1)`**:
   Esta compuerta act√∫a solo si qubit 0 es `1`:
   - $|01‚ü©$ ‚Üí se queda igual
   - $|11‚ü©$ ‚Üí qubit 1 cambia: $|11‚ü© \rightarrow |10‚ü©$

   Resultado final:
   $$
   |\Psi^+‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© + |10‚ü©)
   $$

---

üìå En la QSphere, se ver√°n dos puntos grandes: $|01‚ü©$ y $|10‚ü©$, con el **mismo color**, indicando que tienen la **misma fase**.

## ‚öôÔ∏è Secci√≥n 7.4 ‚Äì Estado de Bell 4: $|\Psi^+‚ü© = \frac{1}{\sqrt{2}}(|01‚ü© - |10‚ü©)$

Este estado de Bell representa una **superposici√≥n entre los estados** \( |00‚ü© \) **y** \( |11‚ü© \), con una **fase negativa** en el segundo t√©rmino:

$$
|\Phi^-‚ü© = \frac{1}{\sqrt{2}}(|00‚ü© - |11‚ü©)
$$

Para construir este estado en Qiskit, seguimos estos pasos:

- Aplicamos una compuerta **Hadamard al qubit 0** para generar una superposici√≥n.
- Aplicamos una compuerta **CNOT**, con el qubit 0 como control y el qubit 1 como objetivo ‚Üí se crea entrelazamiento.
- Finalmente, aplicamos una compuerta **Z al qubit 0**, que invierte la fase del t√©rmino \( |11‚ü© \), introduciendo el signo negativo.

Este proceso nos lleva al estado final:

$$
|\Phi^-‚ü©
$$

Uno de los cuatro **estados de Bell**, fundamentales en protocolos como la **teleportaci√≥n cu√°ntica**, **criptograf√≠a cu√°ntica** y **pruebas de Bell**.

---



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito cu√°ntico con 2 qubits
qc = QuantumCircuit(2)

# Paso 1: Hadamard en qubit 0 ‚Üí superposici√≥n
qc.h(0)

# Paso 2: CNOT entre qubit 0 (control) y qubit 1 (target) ‚Üí entrelazamiento
qc.cx(0, 1)

# Paso 3: Z en qubit 0 ‚Üí cambia la fase del t√©rmino |11‚ü©
qc.z(0)

# Obtener estado
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Notaci√≥n ket:")
display(psi.draw('latex'))

# Visualizar QSphere
print("üß† Visualizaci√≥n QSphere:")
plot_state_qsphere(psi)
````

## üß™ Explicaci√≥n paso a paso del circuito

1. Aplicamos una compuerta **Hadamard al qubit 0**: esto genera una superposici√≥n del tipo

$$
\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
$$

Lo que significa que el sistema queda en:

$$
\frac{1}{\sqrt{2}}(|00\rangle + |10\rangle)
$$

2. Aplicamos una compuerta **CNOT**, con qubit 0 como control y qubit 1 como objetivo. Esto transforma el estado a:

$$
\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)
$$

Ya que la compuerta CNOT cambia el qubit objetivo **solo si** el qubit control est√° en

$$
|1\rangle
$$

3. Aplicamos una compuerta **Z al qubit 0**, que invierte el signo de las componentes donde el qubit 0 est√° en el estado

$$
|1\rangle
$$

En este caso, afecta el t√©rmino

$$
|11\rangle
$$

y el estado final queda como:

$$
|\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle)
$$

---

Este estado es un ejemplo cl√°sico de **entrelazamiento con fase relativa negativa**, lo cual tiene efectos importantes en algoritmos cu√°nticos y protocolos como teleportaci√≥n o criptograf√≠a cu√°ntica.

## üìä 8. Comparaci√≥n visual de los Estados de Bell

Los **estados de Bell** representan las 4 posibles formas de entrelazamiento m√°ximo entre dos qubits. Cada uno es una superposici√≥n de dos estados base, con o sin diferencia de fase.

| Estado de Bell | Expresi√≥n                                                        | Descripci√≥n detallada                                                                                       |
|----------------|------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------|
| **Œ¶‚Å∫**         | (1 / ‚àö2) ¬∑ (  \|00‚ü© + \|11‚ü© )                                     | Ambos qubits est√°n en el mismo estado. Fase positiva.                                                       |
| **Œ¶‚Åª**         | (1 / ‚àö2) ¬∑ (  \|00‚ü© ‚àí \|11‚ü© )                                     | Igual que Œ¶‚Å∫, pero con una fase negativa. El signo menos afecta al t√©rmino \|11‚ü©.                            |
| **Œ®‚Å∫**         | (1 / ‚àö2) ¬∑ (  \|01‚ü© + \|10‚ü© )                                     | Qubits en estados opuestos. Hay simetr√≠a, con fase positiva.                                                |
| **Œ®‚Åª**         | (1 / ‚àö2) ¬∑ (  \|01‚ü© ‚àí \|10‚ü© )                                     | Qubits en estados opuestos, pero con fase negativa que invierte el t√©rmino \|10‚ü©.                            |

---

üìå **Resumen**:
- Los estados `Œ¶` son **correlacionados** (`|00‚ü©`, `|11‚ü©`), mientras que los `Œ®` son **anticorrelacionados** (`|01‚ü©`, `|10‚ü©`)
- Los signos **+ o ‚àí** indican **fase relativa**: afectan la interferencia cu√°ntica
- Todos son **m√°ximamente entrelazados** y **ortogonales entre s√≠**

---



````python

#Codigo para visualizar los estados de bel (ilustrativo)
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
import matplotlib.pyplot as plt
def mostrar_bell(nombre, circuito):
    estado = Statevector.from_instruction(circuito)
    
    print(f"üîé Estado de Bell: {nombre}")
    display(circuito.draw('mpl'))
    display(plot_state_qsphere(estado))
# Œ¶‚Å∫ = (|00‚ü© + |11‚ü©) / ‚àö2
phi_plus = QuantumCircuit(2)
phi_plus.h(0)
phi_plus.cx(0, 1)

# Œ¶‚Åª = (|00‚ü© - |11‚ü©) / ‚àö2
phi_minus = QuantumCircuit(2)
phi_minus.h(0)
phi_minus.cx(0, 1)
phi_minus.z(0)

# Œ®‚Å∫ = (|01‚ü© + |10‚ü©) / ‚àö2
psi_plus = QuantumCircuit(2)
psi_plus.x(1)
psi_plus.h(0)
psi_plus.cx(0, 1)

# Œ®‚Åª = (|01‚ü© - |10‚ü©) / ‚àö2
psi_minus = QuantumCircuit(2)
psi_minus.x(1)
psi_minus.h(0)
psi_minus.cx(0, 1)
psi_minus.z(0)
mostrar_bell("Œ¶‚Å∫ = (|00‚ü© + |11‚ü©) / ‚àö2", phi_plus)
mostrar_bell("Œ¶‚Åª = (|00‚ü© - |11‚ü©) / ‚àö2", phi_minus)
mostrar_bell("Œ®‚Å∫ = (|01‚ü© + |10‚ü©) / ‚àö2", psi_plus)
mostrar_bell("Œ®‚Åª = (|01‚ü© - |10‚ü©) / ‚àö2", psi_minus)
````
